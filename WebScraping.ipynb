**filpkart acer laptop extraction and to convert the csv file**

```python
import bs4
from bs4 import BeautifulSoup as bs
import requests
link='https://www.flipkart.com/search?q=acer+laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=acer+laptop%7CLaptops&requestId=6aa1e7ff-456d-42bb-8b8a-1286240f7000&as-searchtext=acer'
page = requests.get(link)
page.content
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find_all('div',class_="_4rR01T")
print(name)
```
